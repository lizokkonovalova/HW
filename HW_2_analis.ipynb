```python
log = pd.read_csv('visit_log.csv', sep = ';')
#log.head()
def create_source_type(row):
    if row['traffic_source'] in ['yandex', 'google']:
        return 'organic'
    elif row['traffic_source'] in ['paid', 'email']:
        if row['region'] == 'Russia':
            return 'ad'
        return 'other'
    return row['traffic_source']
```

```python
log['source_type'] = log.apply(create_source_type, axis = 1)
log.head(10)
```

```python
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
ratings['rating_count'] = ratings.groupby('userId')['userId'].transform('count')
#ratings
active_users = ratings[ratings['rating_count']>100].sort_values(by=['rating_count'])
#active_users
user_lifetime = active_users.groupby('userId')['timestamp'].agg(['min','max'])
user_lifetime['lifetime'] = user_lifetime['max'] - user_lifetime['min']
average_user_lifetime = user_lifetime['lifetime'].mean()/24/3600
print(f'Среднее время жизни пользователей, которые поставили > 100 оценок составляет {round(average_user_lifetime)} дня')
```
